```python
import gradio as gr
import joblib
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load the trained model
model = joblib.load("model.pkl")

# If you used LabelEncoder for categorical columns during training
# You need to encode the same way in the interface
encoders = {}
categorical_columns = ['Category', 'Brand', 'PaymentMethod', 'State', 'Country']

# Example: load encoders if saved
# for col in categorical_columns:
#     encoders[col] = joblib.load(f"{col}_encoder.pkl")

# Prediction function
def predict_sales(Category, Brand, PaymentMethod, State, Country,
                  Quantity, UnitPrice, Discount, Tax, ShippingCost,
                  SellerID, OrderStatus_Delivered, OrderStatus_Pending,
                  OrderStatus_Returned, OrderStatus_Shipped):
    
    # Encode categorical features (dummy example, if not saved encoders)
    data = pd.DataFrame([{
        'Category': Category,
        'Brand': Brand,
        'PaymentMethod': PaymentMethod,
        'State': State,
        'Country': Country,
        'Quantity': Quantity,
        'UnitPrice': UnitPrice,
        'Discount': Discount,
        'Tax': Tax,
        'ShippingCost': ShippingCost,
        'SellerID': SellerID,
        'OrderStatus_Delivered': OrderStatus_Delivered,
        'OrderStatus_Pending': OrderStatus_Pending,
        'OrderStatus_Returned': OrderStatus_Returned,
        'OrderStatus_Shipped': OrderStatus_Shipped
    }])
    
    # Encode categorical columns using LabelEncoder
    for col in categorical_columns:
        data[col] = LabelEncoder().fit_transform(data[col])
    
    # Predict
    prediction = model.predict(data)[0]
    return round(prediction, 2)

# Define Gradio inputs
inputs = [
    gr.Text(label='Category'),
    gr.Text(label='Brand'),
    gr.Text(label='PaymentMethod'),
    gr.Text(label='State'),
    gr.Text(label='Country'),
    gr.Number(label='Quantity'),
    gr.Number(label='UnitPrice'),
    gr.Number(label='Discount'),
    gr.Number(label='Tax'),
    gr.Number(label='ShippingCost'),
    gr.Number(label='SellerID'),
    gr.Number(label='OrderStatus_Delivered'),
    gr.Number(label='OrderStatus_Pending'),
    gr.Number(label='OrderStatus_Returned'),
    gr.Number(label='OrderStatus_Shipped')
]

# Create interface
interface = gr.Interface(
    fn=predict_sales,
    inputs=inputs,
    outputs=gr.Number(label='Predicted TotalAmount'),
    title="Amazon Order Total Prediction",
    description="Enter the order details and get predicted TotalAmount"
)

# Launch
interface.launch()
```
